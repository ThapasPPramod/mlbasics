# Thompson Sampling

read more on thompson sampling. its based on bernoulli's principle. the code is similar to ocb, only change in the parameters and also probablistic approach rather than deterministic approach of ocb.

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv('Ads_CTR_Optimisation.csv')
```

## Implementing Thompson Sampling

```
import random
N, d = 10000, 10
ads_selected = []
numbers_of_rewards_0, numbers_of_rewards_1 = [0] * d, [0] * d
total_reward = 0
for n in range(N):
  ad = 0
  max_random = 0
  for i in range(d):
    random_beta = random.betavariate(numbers_of_rewards_1[i] + 1, numbers_of_rewards_0[i] + 1)
    if random_beta > max_random:
      max_random = random_beta
      ad = i
  ads_selected.append(ad)
  reward = dataset.values[n, ad]
  numbers_of_rewards_1[ad] += reward
  numbers_of_rewards_0[ad] += 1 - reward
  total_reward += reward
```

## Visualising the results - Histogram

```
plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
